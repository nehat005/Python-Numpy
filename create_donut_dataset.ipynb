```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
def create_donut(radius, size=1000, distribution='uniform'):
    if distribution=='uniform':
        r = np.random.rand(size) + radius
    else:
        r = np.random.randn(size) + radius

    polar_coordinates = np.linspace(0, 2. * np.pi, size)
    x = r * np.sin(polar_coordinates)
    y = r * np.cos(polar_coordinates)
    return x, y
```

```
fig, ax = plt.subplots(ncols=2, figsize=(9,4))

ax[0].set_title("uniform")
x,y = create_donut(5)
ax[0].plot(x,y, linestyle="", marker="o", markersize="2")
x,y = create_donut(10)
ax[0].plot(x,y, linestyle="", marker="o", markersize="2")

ax[1].set_title("normal")
x,y = create_donut(5, distribution='normal')
ax[1].plot(x,y, linestyle="", marker="o", markersize="2")
x,y = create_donut(10, distribution='normal')
ax[1].plot(x,y, linestyle="", marker="o", markersize="2")

for a in ax:
    a.set_aspect("equal")
plt.show()
```
