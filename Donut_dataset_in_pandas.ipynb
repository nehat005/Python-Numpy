```
import pandas as pd
import numpy as np
```

```
np.random.seed(3)
x1=np.random.randint(low=-5,high=5,size=25)
np.random.seed(5)
x2=np.random.randint(low=-10, high=10, size=25)
np.random.seed(7)
y=np.random.randint(2,size=25)

d={'x1':x1, 'x2':x2, 'y':y}
df=pd.DataFrame(data=d)
# df.reset_index(drop=True)
df.reset_index(drop=True, inplace=True)
df
```

```
def quadratic_features(row):
  # print(int(row['x1']))
  # print(int(row['x2']))
  x1_sq=np.square(int(row['x1']))
  x2_sq=np.square(int(row['x2']))
  # print('X1_sq:',x1_sq)
  return (x1_sq, x2_sq)
```

```
# print(df.apply(quadratic_features, axis=1)[:])


df[['x1_2', 'x2_2']] = df.apply(
    lambda df: np.square((df['x1'], df['x2'])), axis=1, result_type='expand')
df
```
